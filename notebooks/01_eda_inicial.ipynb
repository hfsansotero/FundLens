# EDA Inicial — FundLens

Exploratory analysis of the 6 initial funds after `initial_load.py` is run.

```python
import pandas as pd
import numpy as np
import plotly.express as px
from fundlens.storage.database import get_session, init_db
from fundlens.storage import repository as repo

init_db()
session = get_session()
```

```python
# Load all funds and check row counts
funds = repo.get_all_active_funds(session)
for f in funds:
    df = repo.get_prices_df(session, f.id)
    print(f"{f.ticker}: {len(df)} rows | {df['date'].min()} → {df['date'].max()}")
```
